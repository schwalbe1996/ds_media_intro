## 課題：高周波成分を削るとどうなる？
教科書で学んだように、**画像の周波数は「模様の細かさ」を表し**、ノイズや輪郭のような細かい変化は **高周波成分** になります。
そして、平均値フィルタやガウシアンフィルタによる平滑化（ぼかし）は、**高周波成分を削る処理** でした。

ここでは平滑化フィルタを使わず、**周波数領域で直接、高周波成分をゼロにして**画像を作り直し（ローパスフィルタ）、削る範囲を変えると画像がどう変わるかを確かめます。

離散フーリエ変換の結果（`fftshift` 後）は **中央が高周波成分** に対応します。中央から $\pm s$ の範囲をゼロにして逆フーリエ変換すると、高周波成分を取り除いた画像が得られます。$s$ を大きくするほど削る範囲が広がります。

- $s$ を大きくすると、画像はどう変化するか？（くっきり／ぼやける、のどちらか）
