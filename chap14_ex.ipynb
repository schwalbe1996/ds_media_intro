## ステップ1：回転・縮小した画像をつくる（このセルはそのまま実行）

元画像から「45°回転した画像」と「縦横50%に縮小した画像」をつくります。

```python
img = cv2.imread('sample.png')
h, w = img.shape[:2]

# 45度回転した画像
M = cv2.getRotationMatrix2D((w/2, h/2), 45, 1.0)
rotated = cv2.warpAffine(img, M, (w, h))

# 縦横50%に縮小した画像
scaled = cv2.resize(img, None, fx=0.5, fy=0.5)

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1); plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB));     plt.title('元画像');   plt.axis('off')
plt.subplot(1, 3, 2); plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)); plt.title('45°回転'); plt.axis('off')
plt.subplot(1, 3, 3); plt.imshow(cv2.cvtColor(scaled, cv2.COLOR_BGR2RGB));  plt.title('50%縮小'); plt.axis('off')
plt.show()
```
